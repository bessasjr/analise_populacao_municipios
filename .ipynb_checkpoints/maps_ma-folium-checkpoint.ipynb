```python
import json
import pandas as pd
import geopandas as gpd
import folium
```

```python
dtype ={
    'uf': 'category',
    'id': 'str',
    '2001': 'float32',
    '2002': 'float32',
    '2003': 'float32',
    '2004': 'float32',
    '2005': 'float32',
    '2006': 'float32',
    '2007': 'float32',
    '2008': 'float32',
    '2009': 'float32',
    '2010': 'float32',
    '2011': 'float32',
    '2012': 'float32',
    '2013': 'float32',
    '2014': 'float32',
    '2015': 'float32',
    '2016': 'float32',
    '2017': 'float32',
    '2018': 'float32',
    '2019': 'float32',
    '2020': 'float32'
}

df = pd.read_csv('populacao_3.csv', dtype=dtype, engine='c')
```

```python
# Salvando os dados percentuais
dicionario_rn = {}
for i in df[df.uf == 'RN'].index:
    # if df.loc[i][1] in ['Unaí', 'Paracatu', 'Moema', 'Lagoa da Prata', 'Luz', 'Bom Despacho']:
    perc = f'{((df.loc[i][-1]/df.loc[i][3])-1)*100:.2f}'
    #perc = perc.replace('.', ',')
    #print(perc,'|', df.loc[i][1], '|', df.loc[i][2])
    dicionario_rn[df.loc[i][1]] = [df.loc[i][1], df.loc[i][2], int(df.loc[i][-1]), float(perc)]
        
dicionario_rn
```

```python
df_rn = pd.DataFrame(dicionario_rn.values(),columns=['cidade', 'id', 'pop_2020', 'percentual'])
df_rn
```

```python
# Pegando o geojson
url_geojson = 'json_geographic_brazil/nordeste/rn.json'

geo_rn = gpd.read_file(url_geojson)

geo_rn
```

```python
# Levando os dados da taxa de mortalidade para dentro do objeto geoespacial
geo_rn = pd.merge(
    left = geo_rn,
    right = df_rn.filter(items=['id','pop_2020','percentual']), 
    on='id'
)

geo_rn
```

```python
#%%time
# Criando o mapa base
mapa_folium = folium.Map(
    width = 800, 
    height = 600,
    location = [-5.71, -36.54], 
    zoom_start = 7, 
    tiles = "cartodbpositron"
    #tiles = "OpenStreetMap"
    #tiles = 'Mapbox_Control_Room'
    #tiles=None,
    
)

# Criando a camada Choroplet
mapa = folium.Choropleth(
    geo_data = geo_rn,
    name ='choropleth',
    bins=100,
    data = df_rn,
    columns = ['id','percentual'],
    key_on = 'feature.properties.id',
    fill_color = 'Reds',
    nan_fill_color = 'white',
    #nan_fill_opacity = 0.1,
    fill_opacity = 1.0,
    line_opacity = 0.9,
    line_color = 'black', 
    line_weight = 0.25,
    legend_name = 'Variação percentual da população entre 2001 e 2020',
    highlight = True,
    reset = True,

).add_to(mapa_folium)

# folium.Marker([-43.9431934105, -19.7976070923],
#               popup='Some Other Location',
#               icon=folium.Icon(color='darkblue',icon='circle')
#               ).add_to(mapa_folium)


#Inserindo os rotulos ao passar o mouse
mapa.geojson.add_child(
  folium.features.GeoJsonTooltip(
      fields = ['name','id','pop_2020','percentual'],
      aliases = ['Município:','id:','População em 2020:', 'Percentual:'],
      labels = True,
      localize = True,
      sticky = False,
      style=('background-color: black; color: white; font-family: Arial')
      
  )
)


# Renderizando o mapa
mapa_folium
```
